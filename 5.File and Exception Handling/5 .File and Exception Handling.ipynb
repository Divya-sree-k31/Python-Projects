```python
# Excercise 5
def check_positive_list(numbers):

    for num in numbers:
        if num < 0:
            raise ValueError("List contains negative numbers!")

def main():
    print("Enter a list of integers (comma-separated): ")
    input_list = [int(x) for x in input().split(",")]

    try:
        check_positive_list(input_list)
        print("All numbers in the list are positive.")
    except ValueError as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    main()
```

```python
# Excercise 6
def calculate_average():
    try:
        numbers = input("Please enter a list of integers separated by spaces: ")
        number_list = [int(num) for num in numbers.split()]

        if not number_list:
            raise ValueError("The list is empty. Please enter at least one integer.")

        average = sum(number_list) / len(number_list)
        print(f"The average of the entered integers is: {average}")

    except ValueError as e:
        print(f"Error: {e}")

    except Exception as e:
        print(f"An unexpected error occurred: {e}")

    finally:
        print("The program has finished running.")

if __name__ == "__main__":
    calculate_average()
```

```python
# Q1: Create a list of 5 random numbers and print the list
import random
random_list = [random.randint(1, 100) for _ in range(5)]
print("Original List:", random_list)
```

```python

# Q2: Insert 3 new values to the list and print the updated list
random_list.extend([200, 300, 400])
print("Updated List:", random_list)
```

```python

# Q3: Use a for loop to print each element in the list
print("List Elements:")
for num in random_list:
    print(num)
```

```python
# Q1: Create a dictionary with given keys and values
person = {'name': 'John', 'age': 25, 'address': 'New York'}
print("Dictionary:", person)
```

```python
# Q2: Add a new key-value pair ('phone', '1234567890') to the dictionary
person['phone'] = '1234567890'
print("Updated Dictionary:", person)
```

```python
# Q1: Create a set with values 1, 2, 3, 4, and 5
num_set = {1, 2, 3, 4, 5}
print("Original Set:", num_set)
```

```python
# Q2: Add the value 6 to the set
num_set.add(6)
print("Updated Set:", num_set)
```

```python
# Q3: Remove the value 3 from the set
num_set.discard(3)  # discard avoids error if the value is not present
print("Set after removing 3:", num_set)
```

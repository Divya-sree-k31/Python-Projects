```python
# Exercise 3:
import pandas as pd

fruit_series = pd.Series({'apples': 3, 'bananas': 2, 'oranges': 1})
print("\nOriginal Series:")
print(fruit_series)

fruit_series['pears'] = 4
print("\nSeries after adding 'pears':")
print(fruit_series)
```

```python
# Exercise 4:

import random
names = ['Anu', 'Bob', 'Christy', 'Divya', 'Arun', 'Ajay', 'Sree', 'Hitha', 'Isha', 'Nikhil']
ages = [random.randint(25, 45) for _ in range(10)]
genders = random.choices(['Male', 'Female'], k=10)

df = pd.DataFrame({
    'name': names,
    'age': ages,
    'gender': genders
})
print("\nDataFrame with 10 rows:")
print(df)
```

```python
# Exercise 5:
occupations = ['Programmer', 'Manager', 'Analyst'] * 4
occupations = occupations[:10]

df['occupation'] = occupations
print("\nDataFrame after adding occupation column:")
print(df)
```

```python
# Exercise 6:
filtered_df = df[df['age'] >= 30]
print("\nRows where age >= 30:")
print(filtered_df)
```

```python
# Exercise 7:
df.to_csv('employee_data.csv', index=False)
print("\nSaved dataframe to CSV file")

df_from_csv = pd.read_csv('employee_data.csv')
print("\nDataFrame read from CSV:")
print(df_from_csv)
```

```python
# Exercise 1

# MonthNames.py
# Program to convert month number to month name

# Get input from user
month_num = int(input("Enter the month: "))

# Dictionary to store month numbers and names
months = {
    1: "January",
    2: "February",
    3: "March",
    4: "April",
    5: "May",
    6: "June",
    7: "July",
    8: "August",
    9: "September",
    10: "October",
    11: "November",
    12: "December"
}

# Check if input is valid and print result
if month_num in months:
    print(f"Month {month_num} is {months[month_num]}")
else:
    print("Invalid month number. Please enter a number between 1 and 12.")
```

```python
# Exercise 2

def calculate_ticket_price(age):
    full_price = 6.00
    if age < 16:
        return full_price / 2
    elif age >= 60:
        return full_price / 3
    else:
        return full_price

# Get user input
try:
    age = int(input("Enter your age: "))
    ticket_price = calculate_ticket_price(age)
    print(f"Your ticket costs £{ticket_price:.2f}")
except ValueError:
    print("Invalid input. Please enter a valid age.")
```

```python
# Exercise 3

# BodyMassIndex.py
# Program to calculate BMI and determine weight status

# Get input from user
weight = float(input("Enter your weight in (kg): "))
height = float(input("Enter your height in (m): "))

# Calculate BMI
bmi = weight / (height ** 2)

# Determine weight status
if bmi < 18.5:
    category = "underweight"
elif bmi < 25:
    category = "normal"
elif bmi < 30:
    category = "overweight"
else:
    category = "obese"

# Print results
print(f"Your BMI is: {bmi:.2f}")
print(f'You are in the "{category}" range.')
```

```python
# Excercise 5

# Get number from user
num = int(input("Enter a number: "))

# Calculate factorial
factorial = 1
for i in range(1, num + 1):
    factorial *= i

print(f"The factorial of {num} is {factorial}")
```

```python
# Excercise 6

# Get number from user
num = int(input("Enter a number: "))

# Reverse the number
reversed_num = 0
while num > 0:
    digit = num % 10
    reversed_num = (reversed_num * 10) + digit
    num = num // 10

print(f"Reversed number is: {reversed_num}")
```

```python
# Excercise 7

# Get number from user
num = int(input("Enter a number: "))
limit = int(input("Enter limit for multiples: "))

# Print multiples
print(f"Multiples of {num} up to {limit}:")
for i in range(1, limit + 1):
    print(num * i, end=" ")
print()  # For new line at end
```
